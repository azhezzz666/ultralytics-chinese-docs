```python
import cv2

from ultralytics import solutions

# æ‰“å¼€è§†é¢‘æ–‡ä»¶
cap = cv2.VideoCapture("path/to/video.mp4")
assert cap.isOpened(), "è¯»å–è§†é¢‘æ–‡ä»¶æ—¶å‡ºé”™"

# è·å–è§†é¢‘å±æ€§ï¼šå®½åº¦ã€é«˜åº¦å’Œæ¯ç§’å¸§æ•°ï¼ˆfpsï¼‰
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# åœ¨è§†é¢‘å¸§ä¸­å®šä¹‰çº¿æˆ–æ„Ÿå…´è¶£åŒºåŸŸçš„ç‚¹
line_points = [(20, 400), (1080, 400)]  # çº¿åæ ‡

# åˆå§‹åŒ–è§†é¢‘å†™å…¥å™¨ä»¥ä¿å­˜è¾“å‡ºè§†é¢‘
video_writer = cv2.VideoWriter("object_counting_output.avi", cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))

# ä½¿ç”¨å¯è§†åŒ–é€‰é¡¹å’Œå…¶ä»–å‚æ•°åˆå§‹åŒ–ç›®æ ‡è®¡æ•°å™¨
counter = solutions.ObjectCounter(
    show=True,  # å¤„ç†æ—¶æ˜¾ç¤ºå›¾åƒ
    region=line_points,  # æ„Ÿå…´è¶£åŒºåŸŸç‚¹
    model="yolo11n.pt",  # Ultralytics YOLO11 æ¨¡å‹æ–‡ä»¶
    line_width=2,  # çº¿æ¡å’Œè¾¹ç•Œæ¡†çš„ç²—ç»†
)

# å¾ªç¯å¤„ç†è§†é¢‘å¸§
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("è§†é¢‘å¸§ä¸ºç©ºæˆ–è§†é¢‘å¤„ç†å·²æˆåŠŸå®Œæˆã€‚")
        break

    # ä½¿ç”¨ç›®æ ‡è®¡æ•°å™¨è®¡æ•°å¸§ä¸­çš„ç›®æ ‡å¹¶è·å–æ ‡æ³¨åçš„å›¾åƒ
    results = counter(im0)

    # å°†æ ‡æ³¨åçš„å¸§å†™å…¥è¾“å‡ºè§†é¢‘
    video_writer.write(results.plot_im)

# é‡Šæ”¾è§†é¢‘æ•è·å’Œå†™å…¥å™¨å¯¹è±¡
cap.release()
video_writer.release()

# å…³é—­æ‰€æœ‰ OpenCV çª—å£
cv2.destroyAllWindows()
```
