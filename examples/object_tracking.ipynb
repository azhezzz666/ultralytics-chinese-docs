## Python

Python å®ä¾‹åˆ†å‰²å’Œç›®æ ‡è·Ÿè¸ªç¤ºä¾‹ã€‚

```python
from collections import defaultdict

import cv2

from ultralytics import YOLO
from ultralytics.utils.plotting import Annotator, colors

# ç”¨äºå­˜å‚¨è·Ÿè¸ªå†å²çš„å­—å…¸ï¼Œé»˜è®¤ä¸ºç©ºåˆ—è¡¨
track_history = defaultdict(lambda: [])

# åŠ è½½å…·æœ‰åˆ†å‰²åŠŸèƒ½çš„ YOLO æ¨¡å‹
model = YOLO("yolo11n-seg.pt")

# æ‰“å¼€è§†é¢‘æ–‡ä»¶
cap = cv2.VideoCapture("path/to/video.mp4")

# è·å–è§†é¢‘å±æ€§ï¼šå®½åº¦ã€é«˜åº¦å’Œæ¯ç§’å¸§æ•°
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# åˆå§‹åŒ–è§†é¢‘å†™å…¥å™¨ä»¥ä¿å­˜å…·æœ‰æŒ‡å®šå±æ€§çš„è¾“å‡ºè§†é¢‘
out = cv2.VideoWriter("instance-segmentation-object-tracking.avi", cv2.VideoWriter_fourcc(*"MJPG"), fps, (w, h))

while True:
    # ä»è§†é¢‘è¯»å–ä¸€å¸§
    ret, im0 = cap.read()
    if not ret:
        print("è§†é¢‘å¸§ä¸ºç©ºæˆ–è§†é¢‘å¤„ç†å·²æˆåŠŸå®Œæˆã€‚")
        break

    # åˆ›å»ºæ ‡æ³¨å™¨å¯¹è±¡ä»¥åœ¨å¸§ä¸Šç»˜åˆ¶
    annotator = Annotator(im0, line_width=2)

    # å¯¹å½“å‰å¸§æ‰§è¡Œç›®æ ‡è·Ÿè¸ª
    results = model.track(im0, persist=True)

    # æ£€æŸ¥ç»“æœä¸­æ˜¯å¦å­˜åœ¨è·Ÿè¸ª ID å’Œæ©ç 
    if results[0].boxes.is_track and results[0].masks is not None:
        # æå–æ©ç å’Œè·Ÿè¸ª ID
        masks = results[0].masks.xy
        track_ids = results[0].boxes.id.int().cpu().tolist()

        # ç”¨ç›¸åº”çš„è·Ÿè¸ª ID å’Œé¢œè‰²æ ‡æ³¨æ¯ä¸ªæ©ç 
        for mask, track_id in zip(masks, track_ids):
            annotator.seg_bbox(mask=mask, mask_color=colors(int(track_id), True), label=str(track_id))

    # å°†æ ‡æ³¨åçš„å¸§å†™å…¥è¾“å‡ºè§†é¢‘
    out.write(im0)
    # æ˜¾ç¤ºæ ‡æ³¨åçš„å¸§
    cv2.imshow("instance-segmentation-object-tracking", im0)

    # å¦‚æœæŒ‰ä¸‹ 'q' é”®åˆ™é€€å‡ºå¾ªç¯
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# é‡Šæ”¾è§†é¢‘å†™å…¥å™¨å’Œæ•è·å¯¹è±¡ï¼Œå¹¶å…³é—­æ‰€æœ‰ OpenCV çª—å£
out.release()
cap.release()
cv2.destroyAllWindows()
```
